```python
import cv2
import matplotlib.pyplot as plt

def mostrar_imagenes(imagen_BGR):
    plt.figure(figsize=(12, 6))
    
    plt.subplot(131)
    plt.imshow(cv2.cvtColor(imagen_BGR, cv2.COLOR_BGR2RGB))
    plt.title('Imagen en formato BGR (RGB para visualización)')

    imagen_gris = cv2.cvtColor(imagen_BGR, cv2.COLOR_BGR2GRAY)
    plt.subplot(132)
    plt.imshow(imagen_gris, cmap='gray')
    plt.title('Imagen en escala de grises')

    cv2.imwrite('Bellingol.jpg', imagen_gris)
    print('Imagen en escala de grises guardada como imagen_escala_grises.jpg')

    plt.subplot(133)
    plt.imshow(imagen_gris, cmap='gray')
    plt.title('Imagen en escala de grises')
    
    plt.show()

imagen_BGR = cv2.imread('Bellingol.jpg')
mostrar_imagenes(imagen_BGR)
```
